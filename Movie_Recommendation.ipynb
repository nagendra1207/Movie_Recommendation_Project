```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
def recommend(movie_title):
  index= new_df[new_df['title']==movie_title].index[0]
  distance = sorted(list(enumerate(sim[index])),reverse=True,key=lambda vec:vec[1])
  for i in distance[1:11]:
    print(new_df.iloc[i[0]].title)
```

```python
 recommend('Batman')
```

Better Approach

```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
selected_features = ['genres','overview','runtime','tagline','cast','director']
```

```python
for features in selected_features:
  df[features] = df[features].fillna('')
```

```python
df.info()
```

```python
# combining all the selected features
combined_features = df['genres']+ df['overview']+df['tagline']+df['cast']+df['director']
```

```python
print(combined_features)
```

```python
vec = TfidfVectorizer()
```

```python
feature_vectors = vec.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
sim = cosine_similarity(feature_vectors)
```

```python
sim
```

```python
sim.shape
```

```python
def recommend(movie_name):
  # making list of all movies
  all_movies_list = df['title'].tolist()
  # finding close match using difflib
  close_match = difflib.get_close_matches(movie_name,all_movies_list)
  matched_movie = close_match[0]
  index_of_matched_movie = df[df.title==matched_movie]['index'].values[0]
  similarity_score = list(enumerate(sim[index_of_matched_movie]))
  sorted_similar_movies = sorted(similarity_score, key= lambda x:x[1], reverse=True)
  print("Movies Recommended for you:\n")
  for i in sorted_similar_movies[1:11]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
